```python
hnl=[]
dfccl=[]
rl=[]
trl=[]
ovrl=[]
pdpl=[]
fl=[]
pdl=[]
```

```python
start=0
end=4
for page in range(start,end):
    try:
        hon=driver.find_elements(By.XPATH,'//div[@class="property-name"]/span')
        if hon:
            for i in hon:
                hnl.append(i.text)
        else:
            hnl.append('-')
    except Exception as e:
        print("Error processing hostel names: {e}")
        hnl.append('-')

    try:
        difcc=driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
        if difcc:
            for i in difcc:
                dfccl.append(i.get_attribute('href'))
        else:
            dfccl.append('-')
    except Exception as e:
        print("Error processing distance from city center: {e}")
        dfccl.append('-')

    try:
        ra=driver.find_elements(By.XPATH,'//div[@class="rating-score"]/span')
        if ra:
            for i in ra:
                rl.append(i.get_attribute('title'))
        else:
            rl.append('-')
    except Exception as e:
        print("Error processing ratings: {e}")
        rl.append('-')

    try:
        tor=driver.find_elements(By.XPATH,'//div[@class="review"]/span')
        if tor:
            for i in tor:
                trl.append(i.get_attribute('href'))
        else:
            trl.append('-')
    except Exception as e:
        print("Error processing total reviews: {e}")
        trl.append('-')

    try:
        ovre=driver.find_elements(By.XPATH, '//div[@class="rating-score"]/span[2]')
        if ovre:
            for i in ovre:
                ovrl.append(i.text)
        else:
            ovrl.append('-')
    except Exception as e:
        print("Error processing overall reviews: {e}")
        ovrl.append('-')

    try:
        pdfp=driver.find_elements(By.XPATH,'//div[@class="accommodation-price single-row"]/strong')
        if pdfp:
            for i in pdfp:
                pdpl.append(i.text)
        else:
            pdpl.append('-')
    except Exception as e:
        print("Error processing privates & dorms from price: {e}")
        pdpl.append('-')

    try:
        fa=driver.find_elements(By.XPATH, '//div[@class="property-badges"]/span/div/div/div')
        if fa:
            for i in fa:
                fl.append(i.get_attribute('arialabel'))
        else:
            fl.append('-')
    except Exception as e:
        print("Error processing facilities: {e}")
        fl.append('-')

    try:
        prd=driver.find_elements(By.XPATH, '//div[@class="property-description"]/span')
        if prd:
            for i in prd:
                pdl.append(i.text)
        else:
            pdl.append('-')
    except Exception as e:
        print("Error processing property description: {e}")
        pdl.append('-')

next_button=driver.find_element(By.XPATH,'//a[@class="pill-content page-nav nav-right icon-only"]')
next_button.click()
time.sleep(3)
```

```python
print(hnl,dfccl,rl,trl,ovrl,pdpl,fl,pdl)
```
